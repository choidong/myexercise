```python
def min_max(a,b):
    return (a,b) if a>b else (b,a)
max_,min_=min_max(10,20)
print(max_,min_)
```

```python
def factorial(n):
    if n>1:
        return n*factorial(n-1)
    else:
        return 1
n=factorial(10)
print(n)
```

```python
class Member:
    def __init__(self,no:int,name:str,weight:float)->None:
        self.__no=no
        self.__name=name
        self.__weight=weight
    @property
    def weight(self)->float:
        return self.__weight

    @weight.setter
    def weight(self,weight:float) -> None:
        self.__weight=weight if weight>0.0 else 0.0
```

```python
kim=Member(14,'긱김',64.3)
```

```python
print(kim.weight)
```

```python
kim.weight(20)
```

```python
class Commander:
    __count=0
    def __init__(self,name:str)->None:
        self.__name=name
        Commander.__count+=1
        self.__id=Commander.__count
    def id(self)->int:
        return self.__id
    @classmethod
    def max_id(cls)->int:
        return cls.__count
    def print(self)->None:
        print(self.__name,self.__id)
```

```python
kim=Commander('aa')
```

```python
kim.id()
```

```python
Commander.max_id()
```

```python
isinstance(kim,Commander)
```

```python
kim=Commander('ddd')
```

```python
isinstance(kim,Commander)
```
